```python
#categorizing categorical variables

#ENCODINGS GET DUMMIES STARTS HERE


#NOW WE ENCODE

combined1=pd.get_dummies(combined['country'], drop_first=True)

#now we encode

combined2=pd.get_dummies(combined['year'], drop_first=True)

#now we encode

combined3=pd.get_dummies(combined['location_type'], drop_first=True)

combined4=pd.get_dummies(combined['cellphone_access'], drop_first=True)

#now we encode

combined5=pd.get_dummies(combined['gender_of_respondent'], drop_first=True)

#now we encode

combined6=pd.get_dummies(combined['relationship_with_head'], drop_first=True)


combined7=pd.get_dummies(combined['marital_status'], drop_first=True)

#now we encode

combined8=pd.get_dummies(combined['education_level'], drop_first=True)

#now we encode

combined9=pd.get_dummies(combined['job_type'], drop_first=True)
```

```python
#Now we drop the weblog_date column since we have processed it
combined=combined.drop(['country'], axis=1)
combined=combined.drop(['year'], axis=1)

combined=combined.drop(['location_type'], axis=1)
combined=combined.drop(['cellphone_access'], axis=1)
#Now we drop the weblog_date column since we have processed it
combined=combined.drop(['gender_of_respondent'], axis=1)
combined=combined.drop(['relationship_with_head'], axis=1)

combined=combined.drop(['marital_status'], axis=1)
combined=combined.drop(['education_level'], axis=1)
combined=combined.drop(['job_type'], axis=1)
```

```python
#now we combine all datasets together
combined=pd.concat([combined,combined1,combined2,combined3,combined4,combined5,combined6,combined7,combined8,combined9], axis=1)
```

```python
#Now its time to separate those two datasets and we are done with both the train and test set now 
#containing the same number of columns.
train_df=combined[combined['train']==1]
test_df=combined[combined['train']==0]
#we now drop the train columns we created in test and train set
train_df.drop(['train'], axis=1, inplace=True)
test_df.drop(['train'], axis=1, inplace=True)

```
